```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df=pd.read_csv('emails.csv')
df.head()
```

```python
df=df.drop('Email No.',axis=1)
```

```python
X=df.drop('Prediction',axis=1)
y=df['Prediction']
```

```python
#split the data into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#train k-nearest neighbours:KNN
knn=KNeighborsClassifier(n_neighbors=6)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred_knn=knn.predict(X_test)
```

```python
#training with support vector machine
svm=SVC(kernel='linear')
svm.fit(X_train,y_train)
```

```python
y_pred_svm=svm.predict(X_test)
```

```python
#show the results
print("KVM accuracy:",accuracy_score(y_test,y_pred_knn))
print("SVM accuracy:",accuracy_score(y_test,y_pred_svm))
```

```python
print("KNN classification report:",classification_report(y_test,y_pred_knn))
print("SBM classification report:",classification_report(y_test,y_pred_svm))
```
